# 1. Dataset Description

## IBM HR Analytics Employee Attrition & Performance

This dataset contains 1,470 employee records with 35 features, providing insights into factors influencing employee attrition. The dataset includes demographic details, job satisfaction, performance metrics, work-life balance, compensation, and career progression.

### Key features:

* **Target Variable:** Attrition (Yes/No) – Indicates whether an employee has left the company.
* **Demographic Data:** Age, Gender, Marital Status, Education, and Education Field.
* **Job-related Information:** Job Role, Department, Business Travel frequency.
* **Performance & Satisfaction Metrics:** Job Satisfaction, Performance Rating, Work-Life Balance, and Environment Satisfaction.
* **Compensation & Benefits:** Monthly Income, Stock Option Level, Percent Salary Hike.
* **Career Progression:** Years at Company, Years in Current Role, Years Since Last Promotion.

This dataset is commonly used for employee retention analysis, predictive modeling, and HR decision-making to reduce attrition rates and improve employee satisfaction. ​

### 3.3 Mapping

```python
# Mapping dictionaries
education_map = {1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}
env_satisfaction_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
job_involvement_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
job_satisfaction_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
performance_rating_map = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}
relationship_satisfaction_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
work_life_balance_map = {1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best'}

# Applying mappings to the dataframe
df['Education'] = df['Education'].map(education_map)
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].map(env_satisfaction_map)
df['JobInvolvement'] = df['JobInvolvement'].map(job_involvement_map)
df['JobSatisfaction'] = df['JobSatisfaction'].map(job_satisfaction_map)
df['PerformanceRating'] = df['PerformanceRating'].map(performance_rating_map)
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].map(relationship_satisfaction_map)
df['WorkLifeBalance'] = df['WorkLifeBalance'].map(work_life_balance_map)
```

# 4. Analytical Questions

## 4.1 Feature Engineering

```python
# How does attrition vary by tenure category?
df['Tenure Category'] = pd.cut(df['YearsAtCompany'], bins=[0,2,5,10,20], labels=['0-2 years', '3-5 years', '6-10 years', '10+ years'])
df[['Tenure Category','Attrition']]
```

```python
df['Tenure Category'].fillna(df['Tenure Category'].mode()[0], inplace=True)
df['Tenure Category']
```

```python
# How does experience level impact attrition?
df['Experience Level'] = df.apply(lambda x: 'Junior' if x['JobLevel'] in [1,2] else 'Senior', axis=1)
df[['Experience Level', 'Attrition']]
```

```python
# How does total compensation vary by job role and attrition status?
df['Total Compensation'] = df['MonthlyIncome'] * (1 + df['StockOptionLevel'] * 0.05)
df[['Total Compensation', 'Attrition']]
```

```python
# Does salary progression affect attrition?
df['Salary Progression'] = df['MonthlyIncome'] / (df['YearsAtCompany'] + 1)
df[['Salary Progression','Attrition']]
```

```python
# Are employees with longer commute times more likely to leave?
df['Commute Impact'] = df['DistanceFromHome'] * df['YearsAtCompany']
df[['Commute Impact', 'Attrition']]
```

```python
# How does years since last promotion affect attrition?
df['YearsSinceLastPromotion Ratio'] = df['YearsSinceLastPromotion'] / (df['YearsAtCompany'] + 1)
df[['YearsSinceLastPromotion Ratio', 'Attrition']]
```

```python
# Does frequent job role change indicate a higher attrition risk?
df['JobChangeFreq'] = df['TotalWorkingYears'] / (df['NumCompaniesWorked'] + 1)
df[['JobChangeFreq','Attrition']]
```

```python
# What is the effect of managerial support on attrition?
def categorize_manager_support(row):
    if row['RelationshipSatisfaction'] in ['Low', 'Medium'] and row['JobSatisfaction'] in ['Low', 'Medium']:
        return 'Poor Support'
    elif row['RelationshipSatisfaction'] in ['High', 'Very High'] and row['JobSatisfaction'] in ['High', 'Very High']:
        return 'Strong Support'
    else:
        return 'Moderate Support'

df['Manager Support Category'] = df.apply(categorize_manager_support, axis=1)

df[['Manager Support Category','Attrition']]
```
